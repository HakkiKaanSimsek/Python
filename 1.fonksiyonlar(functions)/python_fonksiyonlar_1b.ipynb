```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

df = pd.read_csv('data/flight.csv')
df = df.drop(['tailnum','dep_time','sched_dep_time'],axis=1)
df.head()
```

```python
df['carrier_flight'] = df['carrier'] + df['flight'].astype(str)
df.head()
```

```python
df.insert(loc=0, column='date', value=df.time_hour)
df = df.drop(['time_hour'], axis=1)
df.head()
```

```python
df.groupby('dest').arr_delay.mean().to_frame().sort_values(by='arr_delay',ascending=False)[:10]
```

```python
df.groupby('dest').arr_delay.mean().to_frame().sort_values(by='arr_delay',ascending=False)[:10].sort_index()
```

```python
df[~df.dest.str.startswith(('C','D','G'))][:10]
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek
df.head()
```

```python
week_map={0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
df.day_of_week = df.day_of_week.map(week_map)
df.head()
```
