```python
import pandas as pd
df1 = pd.read_csv('../../projects/airlines/data/flights.csv')
df2 = pd.read_csv('../../projects/airlines/data/airlines.csv')
print(df2.head(3))
df1.head(3)
```

```python
df = pd.merge(df1,df2, left_on='AIRLINE', right_on = 'IATA_CODE')
df.head(3)
```

```python
df.insert(loc=5, column='AIRLINE', value=df.AIRLINE_y)
df.head(3)
```

```python
df.pivot_table(index="AIRLINE",columns="DAY_OF_WEEK",values="ARRIVAL_DELAY")
```

```python
df.pivot_table(index="AIRLINE",columns="DAY_OF_WEEK",values="ARRIVAL_DELAY",aggfunc=lambda x:x.count())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
flight_volume = df.pivot_table(index="AIRLINE",columns="DAY_OF_WEEK",values="DAY",aggfunc=lambda x:x.count())
fv = flight_volume.sort_values(by=1, ascending=False)[:7]
fv = fv.iloc[::-1]
fig = plt.figure(figsize=(16,9))
sns.heatmap(fv, cmap='RdBu',linecolor="w", linewidths=2)
plt.title('Air Traffic Volume by Cities', size=16)
plt.ylabel('CITY')
plt.xticks(rotation=45)
plt.show()
```

```python
pd.concat([df.isnull().sum(), 100 * df.isnull().sum()/len(df)], 
              axis=1).rename(columns={0:'Missing Records', 1:'Percentage (%)'})
```

```python
for i in df.columns:
    if df[i].isnull().sum() > 0.5*len(df):
        df.drop(i, axis = 1, inplace = True)
df.shape
```

```python
for i in df.columns:
    if df[i].nunique() == 1:
        df.drop(i, axis = 1, inplace = True)
df.shape
```
