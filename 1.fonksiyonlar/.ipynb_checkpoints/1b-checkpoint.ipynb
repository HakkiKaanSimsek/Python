```python
import pandas as pd
import numpy as np
%matplotlib inline

df = pd.read_csv('https://raw.githubusercontent.com/HakkiKaanSimsek/Python_Dersleri/master/1.fonksiyonlar/data/flight.csv')
df = df.drop(['tailnum','dep_time','sched_dep_time'],axis=1)
df.head()
```

```python
Concat two columns
```

```python
df['carrier_flight'] = df['carrier'] + df['flight'].astype(str)
df.head()
```

```python
df.insert(loc=0, column='date', value=df.time_hour)
df = df.drop(['time_hour'], axis=1)
df.head()
```

```python
df.groupby('dest').arr_delay.mean().to_frame().sort_values(by='arr_delay',ascending=False)[:10]
```

```python
df.groupby('dest').arr_delay.mean().to_frame().sort_values(by='arr_delay',ascending=False)[:10].sort_index()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek
df.head()
```

```python
week_map={0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
df.day_of_week = df.day_of_week.map(week_map)
df.head()
```

```python
df[~np.isnan(df['arr_delay'])].shape
```

```python
df.loc[:,~df.columns.duplicated()].shape
```

```python
df.drop_duplicates(subset=['origin'], keep='first')
```

```python
pd.concat([df, pd.get_dummies(df['origin'])], axis=1)[:5]
```

```python
pd.concat([df, pd.get_dummies(df['origin'],drop_first=True)], axis=1)[:5]
```

```python
pd.concat([df, pd.get_dummies(df.select_dtypes(include='object'))], axis=1)[:5]
```

```python
df = df[~df.carrier.str.startswith(('9','A','U','D','F','M','V','Y','O'))]
```

```python
df.pivot_table(index="carrier",columns="day_of_week",values="arr_delay")
```

```python
df.pivot_table(index=["origin","carrier"],columns="day_of_week",values="arr_delay",margins=True)
```
